```python
numeros = []
i = 0
while i < 10:
  numeros.append(int(input(f'Digite {10-i} numeros:')))
  i += 1
impares = {x for x in numeros if x % 2 != 0}
pares = {x for x in numeros if x % 2 == 0}
print(f'Impares: {impares}\nPares: {pares}')
```

```python
#primeiro-ultimo
fim = False
lista = []
while fim != True:
  resposta = ""
  lista.append(int(input("Digite um número:")))
  while resposta != "n" and resposta != "s":
    resposta = str(input("Quer finalizar a lista?\n[s]\t[n]").lower())
  if resposta == "s":
    fim = True

#trocando posicoes
original = lista.copy()
temp = lista[0]
lista[0] = lista[-1]
lista[-1] = temp
print(f"Lista original: {original}\nLista modificada: {lista}")
```

```python
#O range mais externo controla o fluxo inteiro, em suma.
#Ele eh (nao tem acento) responsavel por ditar quantas listas serao feitas dentro da lista principal.
#O range mais interno eh repetido nas quantidades dentro dos parenteses, resetado, e repetido tudo do zero de acordo com as repeticoes determinadas no range exterior.
#Logo, o i ira repetir valores durante a execucao, diferentemente do j.
lista = [[1 + j*3 + i for i in range(3)] for j in range(3)] #Agrupamentos de 3 em 3, 3 vezes.
print(lista)
```

```python
produtos = {
    'camiseta': (25.00, 10),
    'calca': (45.00, 5),
    'sapato': (60.00, 3),
    'bone': (15.00, 8)
}
max = float(input("Qual eh o preco maximo?"))
for produto in produtos:
  if produtos[produto][0] < max:
    print(produto)
```

```python
notas = []
contador = 0
ginasta = str(input('Quem eh o ginasta?'))
while contador < 7:
  notas.append(float(input(f'Nota do jurado {contador+1}:')))
  contador += 1
ordem = notas.copy()
ordem.sort()
notas.remove(ordem[0])
notas.remove(ordem[-1])
media = sum(notas)/5
print(f'A media de {ginasta} eh: {media}.')
```
